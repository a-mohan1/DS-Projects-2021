### 6. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
# your code here
# create a copy of item
gen_yr = item.copy(deep=True)
# drop unnecessary columns
gen_yr.drop(['movie id', 'movie title', 'release date'], axis=1, inplace=True)
# group by release year
gen_yr = gen_yr.groupby(['release year']).sum()
# calculate total number of movies by summing over genres
gen_yr['total'] = gen_yr.sum(axis=1)
# get genre as percentage of number of releases in a particular year
gen_yr = gen_yr.div(gen_yr['total'], axis=0) * 100
gen_yr.drop(['total'], axis=1, inplace=True)
gen_yr.head(5)
plt.figure(figsize=(10,6))
sns.heatmap(gen_yr)
```

5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# Group by data according to movie id and use metric count and mean to get count of ratings and average ratings for every movie.
#  the output is provided
movie = data.groupby('movie id')['rating'].agg(['count', 'mean'])
movie.columns = ['count', 'average rating']

movie.head()
```

```python
# Slicing movies have count of ratings more than 100 and storing it in the original 'movie' dataframe. 

movie = movie[movie['count'] > 100]

movie.head()
```

```python
# Sorting the dataframe according to the average movie ratings
movie = movie.sort_values(by='average rating', ascending=False)

movie.head()
```

```python
# Merging with 'item' dataset to get movie titles
# Selecting top 25 movies only

movie.merge(item[['movie id', 'movie title']], on='movie id', how='left').head(25)
```

### 8. Check for the validity of the below statements with respect to the data provided - 9 marks

Considering that all the users in the dataset who have watched a movie have also rated it,
* Genre drama is more popular among men women.  (3 marks)
* Genre sci-fi is more popular among women than men.  (3 marks)
* Genre romance is more popular among men than women  (3 marks)


- Hint: You need to find the fraction of men and women who have rated for the above genres and compare the percentages.

**Please pay attention to what should be the denominator while calculating percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#your code here

# merge the data frames and drop unnecessary columns
df = data.merge(user, on='user id', how='left')
df = df.merge(item, on='movie id', how='left')
df.drop(['user id', 'movie id', 'rating', 'timestamp', 'age', 'release year',
        'occupation', 'zip code', 'movie title', 'release date'], axis=1, inplace=True)

# group by gender and calculate total number of ratings given by men and women for each genre
dfg = df.groupby(['gender']).sum()

# calculate total number of ratings given by men and women for all genres combined
dfg['total'] = dfg.sum(axis=1)

# calculate the percentage of ratings given by men and women for each genre
dfg = dfg.div(dfg['total'], axis=0)*100
dfg.head()
```

**Conclusion:**
* Genre Drama is more popular among the women than men. 20.5% of the women rated Drama movies, but only 18.18% of the men rated Drama movies.
* Genre SciFi is more popular among the men than women. 4.9% of the women rated SciFi movies, and in comparison, 6.36% of the men rated SciFi movies.
* Genre Romance is more popular among the women than men. 10.91% of the women rated Romance movies, but only 8.56% of the men rated Drama movies.
